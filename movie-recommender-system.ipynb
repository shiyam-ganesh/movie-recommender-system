```python
import numpy as np
import pandas as pd
```

```python
movies = pd.read_csv("tmdb_5000_movies.csv")
credits = pd.read_csv("tmdb_5000_credits.csv")
```

```python
movies.head(1) # 20 columns
```

```python
credits.head(1) # 4 columns
```

```python
# the two dataframes are joined upon the "Title" column (also possible upon movie_id)

movies = movies.merge(credits, on="title")
movies.head(1)

#movies.shape => 4809 rows and 23 columns
```

```python
'''Important Attributes/Features:

    -genres
    
    -title
    
    -id (to fetch the poster of the movie from the tmdb website)
    
    -keywords (aka tags: important for content based recommender system)
    
    -overview (get content similarity by comparing overview attribute)
    
    -cast
    
    -crew (mainly directors)
'''
movies.info()
```

```python
movies = movies[['id','title','overview','genres','keywords','cast','crew']]
```

```python
movies.head()
```

```python
movies.isnull().sum()
```

```python
# since overview is very important and there are only 3 null values, we'll drop them.
movies.dropna(inplace=True)
```

```python
# check for any duplicated rows
movies.duplicated().sum()
```

```python
# for the final dataset, we'll have 3 colums: ID, Title and Tags.

'''
Tags contain:
    overview
    genres
    the top 3 cast members
    only the director from crew
''' 
```

```python
movies.iloc[0].genres
```

```python
# as we can notice, the genres is a list of dictionaries, but the list itself is a string.
# To convert it to a list, we use a module called ast.
import ast

# Now we create a helper function to convert the genres to a list of strings that only contain the genre names.
def convert(obj):
    L = []
    for i in ast.literal_eval(obj):
        L.append(i['name'])
    return L
```

```python
movies['genres'] = movies['genres'].apply(convert)
```

```python
movies.head() # notice the genres column.
```

```python
# repeat the same procedure for keywords.
movies['keywords'] = movies['keywords'].apply(convert)
```

```python
movies.head() # notice the keywords column.
```

```python
# pick the top 3 cast members for each movie.
def convert_cast_top3(obj):
    L = []
    counter = 0
    for i in ast.literal_eval(obj):
        if counter != 3:
            L.append(i['name'])
            counter+=1
        else:
            break
    return L
```

```python
movies['cast'] = movies['cast'].apply(convert_cast_top3)
```

```python
movies.head(
```

```python
# function to fetch the director name from the crew members.
def fetch_director(obj):
    L = []
    for i in ast.literal_eval(obj):
        if i['job'] == 'Director' :
            L.append(i['name'])
            break # stop with one director
    return L
```

```python
movies['crew'] = movies['crew'].apply(fetch_director)
```

```python
# we'll convert the text of overview columns to a list of strings.
movies['overview'] = movies['overview'].apply(lambda x: x.split())
```

```python
movies.head()
```

```python
# Now we apply a transformation, which is important in recommender systems.
# We try to remove the spaces, if present, in a single element of the list. For example: Science Fiction => ScienceFiction
movies['genres'] = movies['genres'].apply(lambda x: [i.replace(" ","") for i in x])
movies['keywords'] = movies['keywords'].apply(lambda x: [i.replace(" ","") for i in x])
movies['cast'] = movies['cast'].apply(lambda x: [i.replace(" ","") for i in x])
movies['crew'] = movies['crew'].apply(lambda x: [i.replace(" ","") for i in x])
```

```python
# Now, we concatenate overview, genres, keywords, cast and crew into a single list, called Tags
movies['tags'] = movies['overview'] + movies['genres'] + movies['keywords'] + movies['cast'] + movies['crew']
```

```python
new_df = movies[['id','title','tags']]
```

```python
new_df['tags'] = new_df['tags'].apply(lambda x:" ".join(x))
```

```python
new_df['tags'][0]
```

```python
# convert to lower case (recommended)
new_df['tags'] = new_df['tags'].apply(lambda x: x.lower())
```

```python
new_df.head()
```

```python
# to change similar words to their root words, we use 'stemming'. For example: dance,dancing,danced => danc
# nltk library has this functionality. It is a powerful NLP library.
from nltk.stem.porter import PorterStemmer
ps = PorterStemmer()
```

```python
def stem(text):
    y = []
    for i in text.split():
        y.append(ps.stem(i))
    return " ".join(y)
```

```python
new_df['tags'] = new_df['tags'].apply(stem)
```

```python
# CONVERT THE TAGS TO VECTORS - TEXT VECTORISATION
# Methods Used: bag of words, tfidf, word2vec, etc.
# We will use bag of words.
# Here, we will combine all the 'tags' rows into a large text (corpus) and pick 5000 most occuring words (excluding stop words like is,was,are,there,etc.)
# Then, we find how many times each of the words occur in every movie. This will be a 5000 X 5000 table (5k movies, 5k words), where each row is a vector for a movie.
# The vectors are plotted in a 5000 dimensional space.
# When a vector is picked, the closest 5 vectors are displayed as recommendations.
```

```python
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(max_features=5000, stop_words='english') # pick the top 5000 features and remove stop words
```

```python
vectors = cv.fit_transform(new_df['tags']).toarray() # convert the sparse matrix output to an array
```

```python
cv.get_feature_names()
```

```python
# Now we calculate the cosine similarity of all the vectors. It is nothing but the angle between the vectors. Lesser is the angle, more is the similarity.
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity = cosine_similarity(vectors)
similarity.shape
# similarity of each movie with every other movie is calculated (similarity is just the inverse of cosine distance)
```

```python
list(enumerate(similarity[0])) # to save the indexes of the movies
```

```python
# Now, we have to display 5 similar movies for a given movie.
# For this, first we obtain the index of the movie, then obtain the distances of that movie from similarity matrix using the index
# Then we sort the distances (enumerate is used to save the indexes of the movies after sorting)
# Then we extract the top 5 elements.

sorted(list(enumerate(similarity[0])),reverse=True,key=lambda x: x[1])[1:6] # 5 most similar movies
```

```python
# To extract the index of a movie:
new_df[new_df['title'] == 'The Godfather'].index[0]
```

```python
# Now we define the function to do all this stuff
def recommend(movie):
    
     # obtain the index of the movie
    movie_index = new_df[new_df['title'] == movie].index[0]
    
    # obtain the distances of that movie from similarity matrix
    distances = similarity[movie_index] 
    
    # sort the distances and obtain the 5 most similar movies
    movie_list = sorted(list(enumerate(distances)),reverse=True,key=lambda x: x[1])[1:6]
    
    for i in movie_list:
        print(new_df.iloc[i[0]].title) # i[0] is the index of the movie
        
    return 
```

```python
recommend('Django Unchained')
```
